```python
import numpy as np
import itertools
import pandas as pd
```

```python
def generate_one_hot(probabilities):
    probabilities = np.array(probabilities)
    p_index = np.random.choice(len(probabilities), p=probabilities)
    one_hot = np.zeros_like(probabilities)
    one_hot[p_index] = 1
    return one_hot

product_set = [0, 1, 2, 3]
offer_set = [0, 1, 2, 3]
X = []
Y = []
np.random.seed(10)

#11 = number of possible subsets of the offer set with 2 or more products
hypothetical_choice_p = [[0.98, 0.02, 0, 0],
                         [0.5, 0, 0.5, 0],
                         [0.5, 0, 0, 0.5],
                         [0, 0.5, 0.5, 0],
                         [0, 0.5, 0, 0.5],
                         [0, 0, 0.9, 0.1],
                         [0.49, 0.01, 0.5, 0],
                         [0.49, 0.01, 0, 0.5],
                         [0.5, 0, 0.45, 0.05],
                         [0, 0.5, 0.45, 0.05],
                         [0.49, 0.01, 0.45, 0.05]]

index = 0
for r in range(2, len(offer_set) + 1):
    print(f"r is: {r}")
    for subset in itertools.combinations(offer_set, r):
        print(f"subset is: {subset}")
        print(f"index is: {index}")
        binary_subset = [1 if x in subset else 0 for x in offer_set]
        print(f"binary_subset is: {binary_subset}")
        p = hypothetical_choice_p[index]
        print(f"p is: {p}")
        for _ in range(2):
            print(f"generated example num: {_}")
            X.append(binary_subset)
            print(f"X is: {binary_subset}")
            y = generate_one_hot(p).reshape((1, len(product_set)))
            Y.append(y)
            print(f"Y is: {y}")
        index += 1
```
